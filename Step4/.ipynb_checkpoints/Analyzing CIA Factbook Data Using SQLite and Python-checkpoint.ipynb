## Conencting to the database

```python
import sqlite3
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')

conn = sqlite3.connect("factbook.db")
```

```python
q = "SELECT * from sqlite_master WHERE type = 'table';"
pd.read_sql_query(q,conn)
```

```python
q = "SELECT * from facts LIMIT 5"
pd.read_sql_query(q,conn)
```

## Outliers 

```python
q_min = "select * from facts where population = 0"
pd.read_sql_query(q_min,conn)
```

```python
q_min = "select * from facts where population = (select max(population) from facts)"
pd.read_sql_query(q_min,conn)
```

* The population is 0 in Antarctica which is totally predictable since the area_water is none on this land.  

```python
query_population = "select min(population) min_population ,max(population) max_population,\
            min(population_growth) min_population_growth,max(population_growth) max_population_growth from facts"
pd.read_sql_query(query_population,conn)
```

## Histograms 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
fig = plt.figure(figsize=(15,15))

query_without_min_max = "select population, population_growth, birth_rate,\
         death_rate from facts where (population > (select min(population)\
         from facts)) and (population < (select max(population) from facts)) "
pd.read_sql_query(query_without_min_max,conn).hist(ax = fig.add_subplot(111))
```

## Population density 

```python
q_density = '''select name, cast(population/area_land as float) \
population_density from facts WHERE 
population != (select max(population) from facts)
and population != (select min(population) from facts) 
order by population_density desc;
'''
pd.read_sql_query(q_density,conn)
```
