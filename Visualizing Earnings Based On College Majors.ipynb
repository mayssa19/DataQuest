* The most popular major is business, however, in the above swarmplot, students in the engineering field make the more money. Hence, students in more popular majors does not necessarly earn more money.

## Earnings based on gender 

```python
Major_gender = recent_grads[["Major_category", "Men","Women","Median"]].groupby(['Major_category'], as_index=False).sum()
Major_men = Major_gender[['Major_category',"Median"]][Major_gender['Men'] > Major_gender['Women']]
```

```python
Major_women = Major_gender[['Major_category',"Median"]][Major_gender['Men'] < Major_gender['Women']]
```

```python
g = sns.swarmplot(x="Major_category", y="Median", data=Major_men);
_, labels = plt.xticks()
g.set_xticklabels(labels, rotation=90)
g.set_title("earnings of Major_men")
```

```python
g = sns.swarmplot(x="Major_category", y="Median", data=Major_women);
_, labels = plt.xticks()
g.set_xticklabels(labels, rotation=90)
g.set_title("earnings of Major_women")
```

 * The second plot in this section describes the earnings in majors where the majority are females, however, they don't make more money compared to the plot showing the earnings in majors where the majority are men.
