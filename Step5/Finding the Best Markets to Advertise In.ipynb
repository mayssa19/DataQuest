## The money  for learning 

* In the coming work, we'll be only interested in the top four markets which are USA, India, United Kingdom and Canada since they present the highest frequencies in the frequency distribution table.

* After this analysis, it's clear that USA is a top market to invest in since there are a lot of students who are willing to pay to learn coding.  
* The question that's still rising is which one of the other countries can be a potential market ? At first look, Canda seems the best fit with an amount of money of $39 which is the closest to \$59. However, we should take into account the variablity of each distribution.  
* In order to answer our question, we will use z-scores to figure out which of these three markets \$59 represents the least expensive.

##  Computing  z-scores to find the second best market 

```python
three_coutnries = ['India','United Kingdom','Canada']
for country in three_coutnries:
    mean_country = top_four_countries['MoneyPerMonth'][top_four_countries['CountryLive']==country].mean()
    std_country = top_four_countries['MoneyPerMonth'][top_four_countries['CountryLive']==country].std(ddof=1)
    z = (59 - mean_country) / std_country
    print(country + ': ' + str(z))
```

* The market with the lowest z-score is canada with a value of 0.22. However, India is slightly greater (0.23) which makes the decision harder to choose between them.
